```python
import matplotlib.image as image
import matplotlib.pyplot as plt
import numpy as np
```

```python
class kmeans:
    """Apply kmeans algorithm"""
    def __init__(self, k, iterations=100):
        #initialize number of clusters
        self.k = k
        self.iterations = iterations
    
    def initalize_centroids(self, X):
        #select centroids randomly from our sample
        i = np.random.permutation(X.shape[0])
        centroids = X[i[:self.k]]
        return centroids
        
    def compute_centroid(self, X, labels):
        #update centroids to means of clusters
        centroids = np.zeros((self.k, X.shape[1]))
        for i in range(self.k):
            centroids[i] = np.mean(X[labels == i], axis=0)
            
        return centroids
    
    def compute_distance(self, X, centroids):
        #find distance between datapoints and centroids
        distances = np.zeros((X.shape[0], self.k))
        for i in range(self.k):
            dist = np.linalg.norm(X - centroids[i], axis=1)
            distances[:,i] = np.square(dist)
            
        return distances
    
    def find_closest_cluster(self, distance):
        #find closest cluster for each datapoint
        return np.argmin(distance, axis=1)
    
    def fit(self, X):
        #kmeans algorithm
        self.centroids = self.initalize_centroids(X)
        
        for i in range(self.iterations):
            old_centroids = self.centroids
            distance = self.compute_distance(X, old_centroids)
            self.labels = self.find_closest_cluster(distance)
            self.centroids = self.compute_centroid(X, self.labels)
        
            if np.all(old_centroids == self.centroids):
                break
```

```python
#get dataset
img= image.imread('peppers.bmp')
plt.imshow(img)
#Check shape and dimension
print("Shape: ", img.shape)
print("Dimension: ", img.ndim)


#Reshape this into a matrix that is N  3, where N = x  y is the number of pixels.
pixels = img.reshape((-1,3))
pixels = np.float32(pixels)
print("N*3 shape: ", pixels.shape)
```

```python
#initialise kmeans class with k = 2
kmodel = kmeans(k = 2, iterations = 100) 

kmodel.fit(pixels)

centroids=np.uint8(kmodel.centroids)

#compute labels
predict = kmodel.labels

#replace rgb values of each pixel with the values of the appropriate cluster centre. Draw the final image
label = predict.flatten()
segmented=centroids[label.flatten()]
segmented = segmented.reshape(img.shape)
plt.imshow(segmented)
```
